```python
# Add to make sure plots are displayed inline
%matplotlib inline 
```

```python
# Import dependencies 
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y_axis to a list of floars as the total fare in US dollars accumulated for each month.
y_axis = [ 10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a plot.
plt.plot(x_axis, y_axis)
```

```python
# Create a plot with ax.plt().
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt().
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot 
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend 
plt.legend()
```

```python
# Create the plot. 
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend 
plt.legend()
```

```python
# Create the plot 
plt.plot(x_axis, y_axis, color="green", marker='D', linewidth=2, label='Boston')
#Create a tile 
plt.title("Pyber Fare by Month")
# Create labels for the axes 
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Add a y limit
plt.ylim(0, 45)
#Add grid lines
plt.grid()
```

```python
# Create the plot 
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot 
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#Create the plot 
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
# Add a title 
plt.title("PyBer Fare by Month")
# Add axes lables
plt.ylabel("Fare($)")
plt.xlabel("Dates")
#Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
# Add a title. 
ax.set_title("PyBer Fare by Month")
# Add axes labels.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Dates")
# Invert the y-axis data
plt.gca().invert_yaxis()
```

```python
# Create a scatter plot 
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create a scatter plot 
plt.scatter(x_axis, y_axis)
```

```python
# Create a scatter plot 
plt.scatter(y_axis, x_axis, c='red', label= 'Chicago')
# Add a title 
plt.title("PyBer Fare by Month")
# Add the axes labels
plt.xlabel("Fare($)")
plt.ylabel("Dates")
#Add a legend
plt.legend()
# Add a limit
plt.xlim(0,50)
#Invert the y-axis so that Jan is at the top
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Use list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a scatter plot using object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble chart. 
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a bubble chart.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c='skyblue', alpha=0.2, edgecolor='black', lw=2, label='Boston')
# Add a legend.
plt.legend()
# Add a title.
plt.title("PyBer Fare by Month")
# Add axes labels
plt.ylabel("Dates")
plt.xlabel("Fares($)")
# Add a limit
plt.xlim(0,50)
# Invert the y.axis
plt.gca().invert_yaxis()
```

```python
# Create a pie chart 
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Explode the largest percentage.
plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow","yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
# Create a pie chart using object-oriented interface.
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Create a pie chart using object-oriented interface.
plt.subplots(figsize=(8, 8))
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow","yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots()
ax.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, shadow=True, startangle=90, counterclock=False, autopct='%.1f%%')

plt.show()
```
